**Important**: The data used for this project is found here: https://www.nb.no/sprakbanken/ressurskatalog/oai-nb-no-sbr-50/

```python
import tarfile
import pandas as pd

# Extract the tar file and load content of nynorsk- and bokmål-wiki into a dataframe
data_dir = '/data'

with tarfile.open("2019_wikipedia.tar.gz", "r:gz") as tar:
    tar.extractall()
    
nn = pd.read_json(f"{data_dir}/nno.wikipedia.json")
bm = pd.read_json(f"{data_dir}/nob.wikipedia.json")
```

**NB**! 'Wikidata' acts as the ID for connecting the same Wikipedia-article written in differenet languages

```python
nn.loc[nn['wikidata'] == "Q82300"]
```

```python
bm.loc[bm['wikidata'] == "Q82300"]
```

Relevant columns:
- **Wikidata**: ID connecting similar articles in different languages
- **Text**: Text from each Wikipedia-article
- **Title**: Tile of each Wikipedia-article
- **Bytlength**: Size of each Wikipedia-article

```python
bm = bm[["wikidata", "text", "title", "bytelength"]]
nn = nn[["wikidata", "text", "title", "bytelength"]]
```
